```python
import pandas as pd
df=pd.read_csv("Superstore.csv")
df
```

```python
from matplotlib import pyplot as plt
df.groupby(['Discount']).sum().plot(kind='pie', y='Quantity',figsize=(9,13),pctdistance=1.7,labeldistance=1.2)
```

```python
df.groupby(['Category']).sum().plot(kind='pie', y='Sales',figsize=(9,13),pctdistance=1.7,labeldistance=1.2)
```

```python
df.groupby(['Category']).sum().plot(kind='pie', y='Discount',figsize=(9,13),pctdistance=1.7,labeldistance=1.2)
```

```python
import seaborn as sns
states=df.loc[:,["Sub-Category","Sales"]]
plt.figure(figsize=(10,10))
sns.barplot(x="Sub-Category",y="Sales",data=states)
plt.xticks(rotation = 90)
plt.xlabel=("Sub-Category")
plt.ylabel=("SALES")
plt.show()
```

```python
import seaborn as sns
states=df.loc[:,["Quantity","Sales"]]
plt.figure(figsize=(7,7))
sns.kdeplot(x="Quantity",y="Sales",data=states)
plt.xticks(rotation = 90)
plt.xlabel=("Quantity")
plt.ylabel=("SALES")
plt.show()
```
